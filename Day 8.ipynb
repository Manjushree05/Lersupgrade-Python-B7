```python
##Write a decorator function for your taking input for you any kind of functiom you want to build
for example-you make a fibonacci series function,in which your input range is been defined by the decorator program input
```

```python
def inputdecofun(argfun):
    def newfun():
        num=int(input("enter any number-"))
        argfun(num)
   return newfun
```

```python
@inputdecofun
def fibonacci(n):
    a=0
    b=1
    print('1')
    for i in range(0,n):
        c=b+a
        a=b
        b=c
        print(c)
```

```python
For this challenge you need to develop a python program to find in read only mode and try writing something to it and handle the subsequent errors using exceptiom handling
```

```python
try:
  f = open("test.txt", 'r')
    Try:
      data = f.read()
      print(data)
    except IOError as e:
      print(e)
    except ValueError as e:
      print(e)
    except EOFError as e:
      print(e)
    Except:
      print(“unknown error”)
    Finally:
      f.close()
except:
  print("Fiddlesticks! Failed")
finally:
  print("All Done") 
```
