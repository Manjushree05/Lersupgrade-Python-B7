```python
For this challenge,create a bank account class that has two attributes
*ownerName
*Balance
And two methods
*deposit
*withdraw
As an added requirement,withdrawals may not exceed the available balance.
Instantiate your class,make several deposits and withdrawals,and test to make sure the account cant be overdrawn
```

```python
class Bank():
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
        
    def deposit(self, balance):
        self.balance+=  balance
        return self.balance
    
    def withdraw(self,balance):
            if self.balance >= balance:
                self.balance-=balance
            else:
                return "Not Possible only " + str(self.balance) + ' is avaliable'
            return self.balance
```

```python
account1=Bank('account1',1000)
```

```python
account1.deposit(2000)
```

```python
account1.withdraw(200)
```

```python
account1.withdraw(3000)
```

```python
For this challenge,create a cone class that has two aatributes:
    *R=Radius
    *h=Height
```

```python
import math
class cone():
  def __init__(self,r,h):
   self.r=r
   self.h=h
  def volume_fun(self):
    return math.pi*self.r*self.r*(self.h/3)
  def surface_area_fun(self):
      base= math.pi*self.r*self.r
      side=math.pi*self.r*math.sqrt((self.r*self.r)+(self.h*self.h))
      return base+side
  
    
```

```python
cone1=cone(2,3)
```

```python
cone1.volume_fun()
```

```python
cone1.surface_area_fun()
```
